#Removendo variáveis categóricas

```r
#Removendo outliers
df2 = subset(df, rent.amount..R.. <= 15000 || floor != "-")
df2 = select(df, area, rooms, rent.amount..R.., bathroom, parking.spaces, floor, hoa..R.., property.tax..R.., fire.insurance..R.., total..R..)
# df_quartos = select(df, rooms, rent.amount..R..)
df2
```

```r
library("Hmisc")

matrizCorr <- function(cormat, pmat) {
  ut <- upper.tri(cormat)
  df = data.frame(
    linha = rownames(cormat)[row(cormat)[ut]],
    col = rownames(cormat)[col(cormat)[ut]],
    corr  =(cormat)[ut],
    'p-value' = pmat[ut]
    )
    return (df)
}

# corr <- cor(df2)
# corr
corr2 <- rcorr(as.matrix(df2))

matriz = matrizCorr(corr2$r, corr2$P)
matriz
```

#REGRESSAO

```r
print("AREA")
simple.regression <-lm(rent.amount..R..~area , data = df2)
summary(simple.regression)

print("QUARTOS")
simple.regression <-lm(rent.amount..R..~rooms , data = df2)
summary(simple.regression)

print("BANHEIROS")
simple.regression <-lm(rent.amount..R..~bathroom , data = df2)
summary(simple.regression)

print("VAGAS ESTACIONAMENTO")
simple.regression <-lm(rent.amount..R..~parking.spaces , data = df2)
summary(simple.regression)

print("ANDAR")
simple.regression <-lm(rent.amount..R..~floor , data = df2)
summary(simple.regression)

print("IPTU")
simple.regression <-lm(rent.amount..R..~property.tax..R.. , data = df2)
summary(simple.regression)

print("SEGURO INCENDIO")
simple.regression <-lm(rent.amount..R..~fire.insurance..R.. , data = df2)
summary(simple.regression)

print("VALOR TOTAL IMOVEL")
simple.regression <-lm(rent.amount..R..~total..R.. , data = df2)
summary(simple.regression)
```
